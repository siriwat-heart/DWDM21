#K-means

##Generate Data

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=500, centers=3, n_features=2,random_state=6)  # n_features คือ 2 มิติ | Center เป็นกลุ่ม
# Centers คือ แบ่งข้อมูลออกเป็นกี่กลุ่ม
```

##Explore Data

```python
X.shape # X มี 500 จุด
```

```python
from matplotlib import pyplot as plt    # สุ่มออกมา 3 กลุ่ม   #นำเข้าเพื่อใช้วาดกราฟ
plt.scatter(X[:,0],X[:,1])
```

##Clustering

###import

```python
from sklearn.cluster import KMeans   #นำเข้าเพราะเราจะทำการจักกลุ่มคัสเตอร์
```

###Define

```python
kmeans = KMeans(n_clusters=2, random_state=6)  #กำหนดตัวแปร # น่าจะมี 2 กลุ่ม
```

###Fit-Predict

```python
y_clust = kmeans.fit_predict(X)   # fit_transform แปลงให้ dimension คำนวณได้ง่ายๆ | fit คือให้จัดกลุ่มให้
```

```python
y_clust
```

```python
plt.scatter(X[:,0],X[:,1],c=y_clust)   # ผลลัพธ์ออกมาเป็น สี 2 Cluster
```

3 clusters

```python
#define
kmeans = KMeans(n_clusters=3, random_state=6)
#fit-predict
y_clust = kmeans.fit_predict(X)
#plot
plt.scatter(X[:,0],X[:,1],c=y_clust)
```

4 clusters

```python
#define
kmeans = KMeans(n_clusters=4, random_state=6)
#fit-predict
y_clust = kmeans.fit_predict(X)
#plot
plt.scatter(X[:,0],X[:,1],c=y_clust)
```

```python
# define
kmeans = KMeans(n_clusters=4, random_state=150, init = 'random') # เปลี่ยนกลุ่ม
# fit-predict
y_clust = kmeans.fit_predict(X)
# plot
plt.scatter(X[:,0],X[:,1],c=y_clust)
```

5 clusters

```python
#define
kmeans = KMeans(n_clusters=5, random_state=14, init = 'random')    # เปลี่ยนกลุ่ม
#fit-predict
y_clust = kmeans.fit_predict(X)
#plot
plt.scatter(X[:,0],X[:,1],c=y_clust)
```

```python
#define
kmeans = KMeans(n_clusters=5, random_state=6, init = 'random')   # เปลี่ยนกลุ่ม
#fit-predict
y_clust = kmeans.fit_predict(X)
#plot
plt.scatter(X[:,0],X[:,1],c=y_clust)
```

##Exanple Application (Color Quantization)

```python
import cv2   # การโหลดภาพเข้า
import os  #เพิ่มฟังก์ชั่นที่ต้องการใช้เพื่อใช้การภาพ การนำเข้าภาพ และการใช้สี
```

```python
img = cv2.imread('/content/pexels-tranmautritam-2286016.jpg') #นำเข้าภาพจากไฟล์ที่อัพโหลด
```

```python
img.shape #ตรวจสอบขนาดของภาพ ไม่ควรใหญ่เกินไป
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)  #กำหนนดตัวแปร
```

###นับจำนวนสี

```python
import numpy as np
```

```python
np.unique(np.reshape(RGB_img, (RGB_img.shape[0]*RGB_img.shape[1],3)),axis=0).shape #มี 27547 สี
```

###จัดกลุ่มให้เหลือ 16 สี

```python
myquantize = KMeans(n_clusters=16, random_state=6)
```

###แปลงข้อมูลให้อยู่ในรูป row-column

```python
img_row = np.reshape(RGB_img, (RGB_img.shape[0]*RGB_img.shape[1],3))
img_row.shape
```

```python
final_color = myquantize.fit_predict(img_row)
```

###ใช้ centroid เป็นตัวแทนของสี

```python
centroid = myquantize.cluster_centers_
```

```python
centroid
```

###แทนสีคืนลงไป

```python
new_img = np.zeros(RGB_img.shape)
for i in range(RGB_img.shape[0]):
    for j in range(RGB_img.shape[1]):
        new_img[i,j,:] = centroid[myquantize.predict([RGB_img[i,j,:]]),:]
```

```python
plt.imshow(new_img/255)
```

```python
plt.imsave('img.jpg',new_img/255)
```
